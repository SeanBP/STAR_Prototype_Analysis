```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import mplhep as hep
from scipy.optimize import curve_fit
import pandas as pd
import hist
from hist import Hist
import pickle
```

```python
with open('Run5_list.pickle', 'rb') as handle:
    data = pickle.load(handle)
LG = []
time = []
for evt in data:
    time.append(evt['Tstamp_us'][0])
    LGL = evt['LG']
    for LGs in LGL:
        LG.append(LGs)
```

```python
fig = plt.figure( figsize=(8, 6))

h = hist.Hist(hist.axis.Regular(100, 0, 101, name="Ptrig"))
h.fill(LG)

bin_centers = 0.5 * (h.axes[0].edges[:-1] + h.axes[0].edges[1:])
    
datah = h.values()[()]
bins = bin_centers
plt.errorbar(bins,datah,yerr=np.sqrt(datah),fmt='o',ecolor='tab:blue', color='tab:blue', capsize=0, elinewidth=1, markeredgewidth=0,label="LG Data")
plt.xlabel("Energy [ADCs]")
plt.ylabel("Count")
plt.title("All Channels")
```

```python
fig = plt.figure( figsize=(8, 6))

h = hist.Hist(hist.axis.Regular(100, np.min(time)/1000000, np.max(time)/1000000, name="Ptrig"))

h.fill(np.array(time)/1000000)

bin_centers = 0.5 * (h.axes[0].edges[:-1] + h.axes[0].edges[1:])
    
datah = h.values()[()]
bins = bin_centers
plt.errorbar(bins,datah,yerr=np.sqrt(datah),fmt='o',ecolor='tab:blue', color='tab:blue', capsize=0, elinewidth=1, markeredgewidth=0,label="LG Data")
plt.xlabel("Time Stamp [s]")
plt.ylabel("Frequency [Hz]")
```
